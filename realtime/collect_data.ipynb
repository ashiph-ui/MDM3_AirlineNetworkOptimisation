```python
from flights import Flights
from flightschedules import FlightSchedules

# Create a Flight object
flights = Flights()

# Create a FlightSchedules object
flight_schedules = FlightSchedules(flights)
```

```python
# Get the schedule data from the flight schedules
schedule_data = [flight_schedules.flight_schedules[plane].schedule for plane in flight_schedules.flight_schedules]

# Create a DataFrame from the schedule data
import pandas as pd
df = pd.DataFrame(schedule_data)

# rename rows to plane1, plane2, etc.
df.index = [f"plane{i}" for i in range(len(df))]

# sort columns by flight order (flight1, flight2, etc.)
df = df.reindex(sorted(df.columns), axis=1)

# remove tuple wrapped around the dictionaries
df = df.applymap(lambda x: x[0])

df.head()
```

```python
print("Keys for plane in flight:")
print(list(df.flight1.plane0.keys()))
```

```python
print("Items for plane in flight:")
for key, val in df.flight1.plane0.items():
    print(f"{key}: {val}")
```
